```
import itertools
import pandas as pd
from collections import defaultdict
from mlxtend.frequent_patterns import apriori, association_rules
import time
import os

# List of dataset file paths
data_files = [
    "1_Sl4H0BGjPvBXojeiZrzWH1eWTSZ0NNJ.csv",
    "1nHPfJK6lF4NW3p7jbz4revCm-dgVU9ZO.csv",
    "1t7utYPsVSY3J4_P_ldECgUIO0NIs5Bf8.csv",
    "1JIqcX3FE1H6j0MBZc44Nz1maAdsPGqBT.csv",
    "1BBIBZYnZV4VTCs8SkfUS1QLa0Zm2Vjr2.csv"
]
# Load dataset from CSV
def load_dataset(file_path):
    if not os.path.exists(file_path):
        print(f"Error: File {file_path} not found.")
        return []

    df = pd.read_csv(file_path)
    transactions = df.apply(lambda x: set(x.dropna().astype(str)), axis=1).tolist()
    return transactions

# Generate all possible itemsets (brute force)
def generate_itemsets(items, k):
    return list(itertools.combinations(items, k))

# Calculate support for itemsets
def calculate_support(transactions, itemsets):
    support_count = defaultdict(int)
    total_transactions = len(transactions)

    for transaction in transactions:
        for itemset in itemsets:
            if set(itemset).issubset(transaction):
                support_count[itemset] += 1

    support = {itemset: count / total_transactions for itemset, count in support_count.items()}
    return support

# Brute force algorithm
def brute_force_association_rules(transactions, min_support):
    if not transactions:
        return []

    items = set(item for transaction in transactions for item in transaction)
    k = 1
    frequent_itemsets = []

    while True:
        itemsets = generate_itemsets(items, k)
        support = calculate_support(transactions, itemsets)

        # Filter itemsets based on min support
        valid_itemsets = {k: v for k, v in support.items() if v >= min_support}

        if not valid_itemsets:
            break

        frequent_itemsets.append(valid_itemsets)
        k += 1

    return frequent_itemsets

# Use Apriori from MLxtend
def apriori_algorithm(transactions, min_support):
    df = pd.DataFrame(transactions).fillna("")
    df_encoded = df.applymap(lambda x: 1 if pd.notnull(x) else 0)
    frequent_itemsets = apriori(df_encoded, min_support=min_support, use_colnames=True)
    return frequent_itemsets

# Main function
def main():
    for file_path in data_files:
        print(f"Processing dataset: {file_path}")
        transactions = load_dataset(file_path)

        if not transactions:
            continue

        # Run brute force method
        start_time = time.time()
        brute_force_results = brute_force_association_rules(transactions, min_support=0.3)
        brute_force_time = time.time() - start_time

        # Run Apriori algorithm
        start_time = time.time()
        apriori_results = apriori_algorithm(transactions, min_support=0.3)
        apriori_time = time.time() - start_time

        # Compare results
        print("Brute Force Time:", brute_force_time)
        print("Apriori Time:", apriori_time)

if __name__ == "__main__":
    main()
```
